#### ì´ì‚° ì‹œê°„(Discrete-time) ì‚¬ê° í„ìŠ¤(Rectangular Pulse)ì˜ DTFT(ì´ì‚° ì‹œê°„ í‘¸ë¦¬ì— ë³€í™˜)

##### 1. í’€ì´ ê³¼ì •

**ë¬¸ì œ ì •ì˜:**
ì‹ í˜¸ $x[n]$ì€ ë‹¤ìŒê³¼ ê°™ì´ ì •ì˜.
$$
x[n] = \begin{cases}

\end{cases}
$$

**DTFT ì •ì˜:**
ì´ì‚° ì‹œê°„ í‘¸ë¦¬ì— ë³€í™˜ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŒ.
$$X( \Omega) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\Omega n}$$

**í’€ì´:**
ì‹ í˜¸ê°€ ì¡´ì¬í•˜ëŠ” êµ¬ê°„ $-L$ì—ì„œ $L$ê¹Œì§€ë§Œ í•©ì„ êµ¬í•˜ë©´ ë¨.
$$X(\Omega) = \sum_{n=-L}^{L} 1 \cdot e^{-j\Omega n}$$

ì´ ì‹ì„ í’€ì–´ì„œ ì“°ë©´ ë‹¤ìŒê³¼ ê°™ì€ ë“±ë¹„ìˆ˜ì—´ì˜ í•©ì´ ë¨.
$$X(\Omega) = e^{j\Omega L} + e^{j\Omega (L-1)} + \dots + 1 + \dots + e^{-j\Omega L}$$

ë“±ë¹„ìˆ˜ì—´ì˜ í•© ê³µì‹ $\left( \frac{\text{ì²«í•­} - \text{ë§ˆì§€ë§‰í•­} \times \text{ê³µë¹„}}{1 - \text{ê³µë¹„}} \right)$ì„ ì´ìš©í•˜ê±°ë‚˜, ìœ í•œ ë“±ë¹„ìˆ˜ì—´ ê³µì‹ $\frac{a(r^N - 1)}{r - 1}$ì„ ë³€í˜•í•˜ì—¬ ì ìš©í•  ìˆ˜ ìˆìŒ.    ì—¬ê¸°ì„œ ê°€ì¥ ì‘ì€ í•­ì„ ì²« í•­ìœ¼ë¡œ ë³´ë©´ ì‰¬ì›Œì§ ($n=L$ì¼ ë•Œì¸ $e^{-j\Omega L}$ë¶€í„° ì‹œì‘).   
* ì²« í•­($a$): $e^{j\Omega L}$ (ë˜ëŠ” $e^{-j\Omega L}$ ìˆœì„œì— ë”°ë¼ ë‹¤ë¦„, ì—¬ê¸°ì„œëŠ” í‘œì¤€ ìœ ë„ ê³¼ì •ì„ ë”°ë¦„)

ì¼ë°˜ì ì¸ ìœ ë„ ê³¼ì •:
$$X(\Omega) = \frac{e^{j\Omega L} - e^{-j\Omega(L+1)}}{1 - e^{-j\Omega}}$$

ë¶„ì, ë¶„ëª¨ë¥¼ ëŒ€ì¹­ í˜•íƒœë¡œ ë§Œë“¤ê¸° ìœ„í•´ ë¶„ìëŠ” $e^{-j\Omega/2}$, ë¶„ëª¨ë„ $e^{-j\Omega/2}$ë¡œ ë¬¶ì–´ë‚´ê±°ë‚˜, ìœ„ ì‹ì˜ ë¶„ì ë¶„ëª¨ì— $e^{j\Omega/2}$ë¥¼ ê³±í•˜ì—¬ ì •ë¦¬í•˜ë©´,

$$X(\Omega) = \frac{2j \sin(\Omega(L + \frac{1}{2}))}{2j \sin(\frac{\Omega}{2})}$$

$$X(\Omega) = \frac{\sin(\Omega(L + \frac{1}{2}))}{\sin(\frac{\Omega}{2})}$$
ë˜ëŠ” $N = 2L+1$ (í„ìŠ¤ì˜ ì „ì²´ ê¸¸ì´)ì´ë¼ê³  í•  ë•Œ, ë‹¤ìŒê³¼ ê°™ì´ í‘œí˜„í•˜ê¸°ë„ í•¨.
$$X(\Omega) = \frac{\sin(\frac{\Omega N}{2})}{\sin(\frac{\Omega}{2})}$$



````python
import numpy as np
import matplotlib.pyplot as plt

def plot_rectangular_pulse_dtft(L=2):
    """
    L: ì¤‘ì‹¬ 0ì„ ê¸°ì¤€ìœ¼ë¡œ ì–‘ì˜†ìœ¼ë¡œ ë»—ëŠ” ìƒ˜í”Œ ìˆ˜ (ì „ì²´ ê¸¸ì´ N = 2L + 1)
    """
    # 1. ì´ì‚° ì‹ í˜¸ x[n] ìƒì„± (ì‹œê°í™” ìš©)
    n = np.arange(-L - 2, L + 3)
    x = np.where((n >= -L) & (n <= L), 1, 0)

    # 2. ì£¼íŒŒìˆ˜ ì¶• ì„¤ì • (-pi ~ pi)
    omega = np.linspace(-np.pi*3, np.pi*3, 1000)

    # 3. DTFT ê³„ì‚° (ìœ ë„í•œ ê³µì‹ ì‚¬ìš©)
    # ë¶„ëª¨ê°€ 0ì´ ë˜ëŠ” ì§€ì (omega=0)ì—ì„œì˜ NaN ë°©ì§€ë¥¼ ìœ„í•´ ì—¡ì‹¤ë¡ (eps) ì¶”ê°€ ë˜ëŠ” ì˜ˆì™¸ ì²˜ë¦¬
    # L'Hopital ì •ë¦¬ì— ì˜í•´ omega=0 ì¼ ë•Œ ê°’ì€ 2L+1 ì´ ë©ë‹ˆë‹¤.
    eps = 1e-10
    numerator = np.sin(omega * (L + 0.5))
    denominator = np.sin(omega / 2) + eps # 0ìœ¼ë¡œ ë‚˜ëˆ„ê¸° ë°©ì§€

    X_omega = numerator / denominator

    # ---------------- ê·¸ë˜í”„ ê·¸ë¦¬ê¸° ----------------
    plt.figure(figsize=(12, 8))

    # (1) ì‹œê°„ ë„ë©”ì¸ ì‹ í˜¸ x[n]
    plt.subplot(2, 1, 1)
    plt.stem(n, x, basefmt="k")
    plt.title(f'Discrete Time Signal x[n] (L={L})')
    plt.xlabel('n')
    plt.ylabel('x[n]')
    plt.grid(True)
    plt.xlim([-L-2, L+2])
    plt.ylim([0, 1.5])

    # (2) ì£¼íŒŒìˆ˜ ë„ë©”ì¸ DTFT X(e^jw)
    plt.subplot(2, 1, 2)
    plt.plot(omega, X_omega, 'b-', linewidth=2)

    # ì£¼ìš” ì§€ì  í‘œì‹œ
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

    plt.title(r'DTFT Magnitude $X(e^{j\omega}) = \frac{\sin(\omega(L+1/2))}{\sin(\omega/2)}$')
    plt.xlabel(r'Frequency $\Omega$ (rad/sample)')
    plt.ylabel('Amplitude')

    # xì¶• ëˆˆê¸ˆì„ pi ë‹¨ìœ„ë¡œ í‘œì‹œ
    plt.xticks(
        [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
        [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']
    )
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# í•¨ìˆ˜ ì‹¤í–‰ (ì˜ˆ: L=3 ì¼ ë•Œ)
plot_rectangular_pulse_dtft(L=3)
````

````python

# Importing Required Libraries
import numpy as np
from matplotlib import pyplot as plt

# Generating time data using arange function from numpy
time = np.arange(-2*np.pi, 2*np.pi, 0.01)

# Finding amplitude at each time using np.sinc function
#amplitude = np.sinc(time)
amplitude = np.sin( np.pi * time * 1.5) / np.sin( np.pi * time * 0.1)

# Plotting time vs amplitude using plot function from pyplot
plt.plot(time, amplitude)

# Settng title for the plot in blue color
plt.title('Sinc Function', color='b')

# Setting x axis label for the plot
plt.xlabel('Time'+ r'$\rightarrow$')

# Setting y axis label for the plot
plt.ylabel('Sinc(time) '+ r'$\rightarrow$')

# Showing grid
plt.grid()

# Highlighting axis at x=0 and y=0
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

# Finally displaying the plot
plt.show()
````
